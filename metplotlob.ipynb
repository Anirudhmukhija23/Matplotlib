```python
#axes -xaxis, y axis, zaxis,labes,tickers
x=np.linspace(0,10,100)
print(x)
fig, axes= plt.subplots(2,3)
axes[0,0].plot(x,np.sin(x))
axes[0,0].set_title("sin _func")
axes[0,0].set_xlabel("x")
axes[0,0].set_ylabel("y")


axes[0,1].plot(x,np.cos(x))
axes[0,1].set_title("cos _func")
axes[0,1].set_xlabel("x")
axes[0,1].set_ylabel("y")
# zeoth row and second column
axes[0,2].plot(x,np.tan(x))
axes[0,2].set_title("tan _func")
axes[0,2].set_xlabel("x")
axes[0,2].set_ylabel("y")

#one row amd zeroth col
axes[1,0].plot(x,np.exp(x))
axes[1,0].set_title("exp _func")
axes[1,0].set_xlabel("x")
axes[1,0].set_ylabel("y")


axes[1,1].plot(x,np.sqrt(x))
axes[1,1].set_title("sqrt _func")
axes[1,1].set_xlabel("x")
axes[1,1].set_ylabel("y")


axes[1,2].plot(x,x)
axes[1,2].set_title("linaer_func")
axes[1,2].set_xlabel("x")
axes[1,2].set_ylabel("y")

plt.tight_layout()


```

```python
#bar plot =categorical data pe kaam krega
courses=["genAI","HLD","DSA","LLD","CCP"]
students=[2500,3000,3200,2800,2600]

plt.bar(courses,students,width=0.5,align="center",edgecolor="black",linewidth=3,color="red",hatch="*")
plt.show()
```

```python

#diggernt hatches iin each bar
courses=["genAI","HLD","DSA","LLD","CCP"]
students=[2500,3000,3200,2800,2600]
hatches=["/","*","?","-","+"]
bar_colors=["red","green","yellow","m","orange"]

bars= plt.bar(courses,students,width=0.5,align="center",edgecolor="black",linewidth=3,)
for bar,hatch,color in zip(bars,hatches,bar_colors):
    bar.set_hatch(hatch)
    bar.set_color(color)
    bar.set_edgecolor("black")
plt.xlabel("Courses")
plt.ylabel("No of Students")
plt.title("BAR PLOT,NO OF STUDENTS ENROL IN COURESES")
plt.show()
```

```python
#histogram plot=x= continous distribution
ages=[34,23,78,65,43,21,98]
sorted_ages=np.sort(ages)
print(sorted_ages)
plt.xlabel("COUNT")
plt.ylabel("Ages")
plt.title(" HISt PLOT No Of students enrolled in courses")
plt.hist(ages,bins=4,orientation="horizontal",linewidth=2,edgecolor="black",color="red")
```

```python
#pie plot
courses=["genAI","HLD","DSA","LLD","CCP"]
pie_colors=["red","green","yellow","m","orange"]
students=[2500,3000,3200,2800,2600]
explode=[0.2,0,0,0,0]
plt.pie(students,labels=courses,shadow=True ,explode=explode,colors=pie_colors, autopct="%.1f%%")
plt.legend()
plt.show()
```

```python
ax=plt.axes(projection="3d")

x=np.random.randint(1,10,5)
y=np.random.randint(1,10,5)
z=np.zeros(5)
dx=np.ones(5)
dy=np.ones(5)
dz=np.random.randint(1,10,5)
print(x,y,z,dx,dy,dz)



ax.bar3d(x,y,z,dx,dy,dz)
ax.set_xlabel("x")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```

```python
#plot surface
ax=plt.axes(projection="3d")
x=np.arange(-5,5,0.25)
y=np.arange(-5,5,0.25)
print(x,y)
x,y=np.meshgrid(x,y)
R=np.sqrt(x**2+y**2)
Z=np.sin(R)
ax.plot_surface(x,y,Z, cmap="plasma")
```
